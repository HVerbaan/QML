```python
import matplotlib.pyplot as plt

case = 5

# =========================
# CASE DEFINITIONS
# =========================

if case == 1:

    charge_list = [[142.2, 57.8, 950.2, 16.8], [277.6, 12.4], [320.2, 0.6], [669, 21.4]]

    charge_increase = [[52.2, 6.6], [83.8], [75.2], [66.7]]

    y_battery = [
        [110, 90, 57.8, 49.1, 34.9, 28.9, 16.8, 0],
        [110, 79.7, 23.2, 12.4, 86.2, 76.2, 0],
        [110, 51.9, 17.8, 0.6, 60.2, 51.2, 0],
        [110, 43.4, 35.4, 21.4, 78.1, 0]
    ]

    x_time = [
        [0, 20, 142.2, 647.9, 752.1, 848.1, 950.2, 1600],
        [0, 30.3, 176.8, 277.6, 934, 1119, 1600],
        [0, 88.9, 213, 320.2, 1073, 1171.9, 1600],
        [0, 311, 409, 669, 842, 1600]
    ]
## Case 2
elif case == 2:

    charge_list = [[142.2, 57.8, 950.2, 16.8], [277.6, 12.4], [320.2, 0.6], [669, 21.4]]

    charge_increase = [[52.2, 6.6], [83.8], [75.2], [66.7]]

    y_battery = [
        [110, 90, 57.8, 49.1, 34.9, 28.9, 16.8, 0],
        [110, 79.7, 23.2, 12.4, 86.2, 76.2, 0],
        [110, 51.9, 17.8, 0.6, 60.2, 51.2, 0],
        [110, 43.4, 35.4, 21.4, 78.1, 0]
    ]

    x_time = [
        [0, 20, 142.2, 647.9, 752.1, 848.1, 950.2, 1600],
        [0, 30.3, 176.8, 277.6, 934, 1119, 1600],
        [0, 88.9, 213, 320.2, 1073, 1171.9, 1600],
        [0, 311, 409, 669, 842, 1600]
    ]
## Case 3
elif case == 3:

    charge_list = [[277.6, 41.7], [407, 59.2], [1221, 16.2], []]

    charge_increase = [[14.4], [18], [19.6], []]

    y_battery = [
        [110, 88.8, 49.2, 41.7, 48.5, 37.6, 27.7, 23.5, 0],
        [110, 63.4, 59.2, 70.2, 60.3, 53.3, 0],
        [110, 69.3, 45.5, 33.4, 22.5, 16.2, 0],
        [110, 96, 73.4, 37.8, 21.5]
    ]

    x_time = [
        [0, 30.3, 176.8, 277.6, 405, 510.6, 705.1, 801.1, 1600],
        [0, 311, 407, 989.9, 1094, 1194, 1600],
        [0, 88.9, 213, 320.2, 1122.1, 1221, 1600],
        [0, 170, 292.2, 887, 1600]
    ]
## Case 4
elif case == 4:

    charge_list = [[], [889.9, 7.2], [179.5, 27.4], [896.2, 10.7]]

    charge_increase = [[], [70], [29.4], [5.7]]

    y_battery = [
        [90, 49.3, 43.5, 37.3, 1.4],
        [90, 76, 53.4, 7.2, 70.2, 60.3, 53.3, 0],
        [90, 34.9, 27.4, 49.2, 43.6, 32.7, 0],
        [90, 68.8, 60.7, 48.6, 23.4, 19.2, 10.7, 0]
    ]

    x_time = [
        [0, 227, 325.2, 1171.9, 1600],
        [0, 20, 142.2, 889.9, 989.9, 1094, 1279, 1600],
        [0, 78.7, 179.5, 316.8, 414.8, 601, 1600],
        [0, 30.3, 131.9, 320.2, 698, 794, 896.2, 1600]
    ]
## Case 5
elif case == 5:

    charge_list = [[], [], [], []]

    charge_increase = [[], [], [], []]

    y_battery = [
        [140, 126, 103.4, 68],
        [140, 99.3, 75.5, 63.4, 52.5, 46.2, 10.4],
        [140, 93.4, 89.2, 82.2, 72.3, 65.3, 12],
        [140, 118.8, 79.2, 71.7, 64.1, 53.2, 43.3, 39.1, 30.6, 14.3]
    ]

    x_time = [
        [0, 20, 142.2, 1600],
        [0, 58.1, 363.8, 471, 1073, 1221, 1600],
        [0, 471, 742, 842, 946.1, 1279, 1445.2],
        [0, 30.3, 176.8, 277.6, 405, 510.6, 848.8, 944.8, 1047, 1600]
    ]



def plotting_battery(x_time, y_battery, case, charge_list, charge_increase):
    import matplotlib.pyplot as plt

    plt.figure(figsize=(8, 5))
    labels = ['Vehicle 1', 'Vehicle 2', 'Vehicle 3', 'Vehicle 4']
    lines = []

    # Plot battery curves
    for i in range(len(y_battery)):
        line, = plt.plot(
            x_time[i],
            y_battery[i],
            marker='o',
            label=labels[i]
        )
        lines.append(line)

    # Plot charge annotations
    for v in range(len(charge_list)):

        if not charge_list[v] or not charge_increase[v]:
            continue

        color = lines[v].get_color()

        # Loop through charge points, only as many as charge_increase values
        for j, inc in enumerate(charge_increase[v]):
            x_charge = charge_list[v][j]

            # Find closest y value on battery curve
            x_arr = x_time[v]
            y_arr = y_battery[v]
            idx = min(range(len(x_arr)), key=lambda k: abs(x_arr[k] - x_charge))
            y_charge = y_arr[idx]

            # New dot at charged value
            y_new = y_charge + inc
            plt.plot(x_charge, y_new, marker='o', color=color, markersize=8)

            # Vertical line from original point to new dot
            plt.vlines(
                x_charge,
                y_charge,
                y_new,
                linestyles='-.',
                colors=color,
                linewidth=2
            )

            # Annotation pointing at the new dot
            plt.annotate(
                f"+{inc:.1f}",
                xy=(x_charge, y_new),
                xytext=(x_charge + 80, y_new + 5),
                arrowprops=dict(arrowstyle="->", color=color),
                bbox=dict(boxstyle="round", fc="white", ec=color, alpha=0.9)
            )

    plt.xlabel("Time unit")
    plt.ylabel("Battery level")
    plt.title("Battery level throughout time")
    plt.legend()
    plt.grid(True)
    plt.savefig(f"case_{case}_battery.png", dpi=300, bbox_inches="tight")
    plt.show()




# =========================
# RUN
# =========================

plotting_battery(x_time, y_battery, case, charge_list, charge_increase)
```
